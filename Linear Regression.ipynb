```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier 
#from sklearn import tree

dfSalary=pd.read_csv('Salary_Data.csv')
inputDataSet = dfSalary.drop(columns=['Salary']) # to define the input data set remove output data set which makes predictions
outputDataSet = dfSalary['Salary']

linreg = LinearRegression()
linreg.fit(inputDataSet,outputDataSet) 
print ("Intercept is : "+ str(linreg.intercept_))
print ("Coefficient is : "+str(linreg.coef_))
feat_cols=['YearsExperience']
zip(feat_cols,linreg.coef_)

inputDataSet_train,inputDataSet_test,outputDataSet_train,outputDataSet_test = train_test_split(inputDataSet, outputDataSet)


modelDC = DecisionTreeClassifier() # create object and set it to new instance

modelDC.fit(inputDataSet_train,outputDataSet_train)  
myPredictions= modelDC.predict(inputDataSet_test) 

myaccuracyscore = accuracy_score(outputDataSet_test,myPredictions) #find your accuracy score
#print ("Accuracy Score is : "+ str(myaccuracyscore))

print("Mean Absolute Error is : " + str(metrics.mean_absolute_error(outputDataSet_test,myPredictions)))
print("Mean Squared Error is : " +str(metrics.mean_squared_error(outputDataSet_test,myPredictions)))
#MSE = np.square(np.subtract(outputDataSet_test,myPredictions)).mean()
#------------Root Mean Squared Error
#RMSE = np.sqrt(MSE)
print ("Root Mean Squared Error is : "+str(np.sqrt(metrics.mean_squared_error(outputDataSet_test,myPredictions))))
dfSalary.head()
#sns.distplot(dfSalary['Salary'])
sns.jointplot(x='YearsExperience', y='Salary', data=dfSalary, kind='reg')
```
