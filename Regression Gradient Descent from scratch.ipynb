# AIML-C6-Regression_Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 8, 10])
y = np.array([5, 7, 9, 11, 13, 18, 20, 22])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def gradient_descent(x,y):
    w0 = 0
    w1 = 0
    iterations = 1000 # epochs
    n = len(x)
    eta = 0.001
    for i in range(iterations):
        y_pred = w0 + w1*x
        cost = (1/n)*sum(val**2 for val in (y-y_pred))
        w0_D = -(2/n)*sum((y-y_pred))
        w1_D = -(2/n)*sum(x*(y-y_pred))
        w0 = w0 - eta * w0_D
        w1 = w1 - eta * w1_D
        print("w0{},w1{},iteration{},cost{}".format(w0,w1,i,cost)) 
        
gradient_descent(x,y)
```

```python
y_pred = 1.513071753770061 + 2.2922714333885965*x
```

```python
plt.scatter(x, y)
plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)],color = 'red')
plt.show()
```
